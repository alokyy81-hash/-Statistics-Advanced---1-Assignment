```python
import pandas as pd
from sklearn.linear_model import LinearRegression
data = pd.read_csv('/content/marketing.csv')
data
x = data[['marketing_spend($)']]
y = data['sales(units)']
model = LinearRegression()
model.fit(x, y)
y_pred = model.predict(x)
print(y_pred)

data['Predict_sales'] = y_pred
y_pred
```

```python
data
```

```python
print("coefficient of the line",model.coef_)
print("y intercept of the line",model.intercept_)
```

```python
import matplotlib.pyplot as plt

plt.scatter(x,y, color='blue')
plt.plot(x,model.coef_*x+model.intercept_, color='red')
```
